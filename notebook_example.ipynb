```python
import requests
from datetime import datetime
import pandas as pd
import time
import json
urlbase = 'https://mighty-bastion-45199.herokuapp.com/'
token = 'token_dummy_001'

def get_result(x):
    try:
        result = pd.DataFrame.from_dict(x.json())
    except:
        result = x.text
    return result

def api_post(route, payload):
    url = urlbase + route
    x = requests.post(url, data = payload)
    df = get_result(x)
    return df

def api_get(route):
    url = urlbase + route
    x = requests.get(url)
    df = get_result(x)
    return df
```
